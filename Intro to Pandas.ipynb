## Intro to Pandas
1. pandas is a software library written for the Python for data manipulation and analysis. 
2. It offers data structures and operations for manipulating tables and time series.
3. Features
    - DataFrame object for data manipulation with integrated indexing.
    - Tools for reading and writing data between in-memory data structures and different file formats.
    - Data alignment and integrated handling of missing data.
    - Reshaping and pivoting of data sets.
    - Label-based slicing, fancy indexing, and subsetting of large data sets.
    - Data structure column insertion and deletion.
    - Group by engine allowing split-apply-combine operations on data sets.
    - Data set merging and joining.
    - Hierarchical axis indexing to work with high-dimensional data in a lower-dimensional data structure.
    - Time series-functionality: Date range generation and frequency conversions, moving   window statistics, moving window linear regressions, date shifting and lagging. Provides data filtration.

(Wikipedia)

```python
# Dataframe of USA covid data
covid19_USA = covid19[covid19['iso_code'] == 'USA']
print(covid19_USA.shape)
covid19_USA.head()
```

```python
# A useful dictionary called Counter
from collections import Counter
mylist = [1,3,4,5,3,2,1,2,4,4,6,7]
c = Counter(mylist)
print(c)
print('unique in mylist:',sorted(c.keys()))
print('Counts:',c.values())
print('Most repeated element:',c.most_common())
print('Most repeated two elements',c.most_common(2))
```

```python
# Sanity check
c = Counter(covid19_USA['location'])
print(c)
```

```python
# Getting rid of redundant fields and reset index
covid19_USA = covid19_USA.drop(['iso_code','continent','location'],axis = 1)
covid19_USA = covid19_USA.reset_index()
covid19_USA = covid19_USA.drop(['index'],axis = 1)
print(covid19_USA.shape)
covid19_USA.head()
```

```python
covid19_USA = covid19_USA[['date','total_cases','new_cases',
                           'new_deaths','new_tests','positive_rate','new_vaccinations']]
print(covid19_USA.shape)
print(covid19_USA.info())
covid19_USA.head()
```

```python
covid19_USA.tail()
```

```python
covid19_USA = covid19_USA.fillna(0)
covid19_USA.head()
```

```python
# One graphic tool
import matplotlib.pyplot as plt

covid19_USA['total_cases'].plot()
```

```python
covid19_USA.set_index('date',inplace=True)
print(covid19_USA.info())
covid19_USA.head()
```

```python
covid19_USA.index_col = 'date'
covid19_USA.head()
```

```python
covid19_USA['total_cases'].plot(figsize=(20, 8))
```

```python
covid19_USA['new_cases'].plot(figsize=(20, 8))
```

```python
csum = covid19_USA['new_cases'].cumsum()
csum.plot(figsize=(20, 8))
```

```python
cases_vaccine = covid19_USA[['new_cases','new_vaccinations']]
cases_vaccine.plot(figsize=(20, 8))
```

```python
covid19_USA['total_vaccinations'] = covid19_USA['new_vaccinations'].cumsum()
cases_vaccine = covid19_USA[['total_cases','total_vaccinations']]
cases_vaccine.plot(figsize=(20, 8))
```

```python
# Z-Score outlier detection
mn = covid19_USA['new_cases'].mean()
std = covid19_USA['new_cases'].std()
covid19_USA['z-score_nc'] = (covid19_USA['new_cases'] - mn)/std
covid19_USA.head()
```

```python
covid_outliers = covid19_USA[(covid19_USA['z-score_nc'] < -3) & 
                             (covid19_USA['z-score_nc'] > 3)]
print(covid_outliers.shape)
```
